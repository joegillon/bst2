```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = "https://www.geographic.org/streetview/usa/mi/ann_arbor.html"
req = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
```

```python
parser=bs(req.text, 'html.parser')
```

```python
span=parser.find('span', class_='listspan')
```

```python
span
```

```python
ul=span.findChild('ul')
```

```python
li=ul.findChildren('li')
li
```

```python
li[0].findChild('a').get_text()
```

```python
streets=[]
for item in li:
    streets.append(item.findChild('a').get_text())
streets
```

```python
import pickle
```

```python
with open('c:/bench/bst/data/mi/streets.txt', 'w') as f:
    for street in streets:
        f.write(street + '\n')
```
