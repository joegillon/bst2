```python
def build_street(parts):
    sname = ''
    for t in parts:
        if t[1] in [
            'StreetNamePreDirectional',
            'StreetName',
            'StreetNamePostType',
            'StreetNamePostDirectional'
        ]:
            sname += t[0] + ' '
    return sname.strip()
```

```python
street=[]
for addr in df.street_address:
    us_addr = usaddress.parse(addr)
    street.append(build_street(us_addr))
len(street)
```
